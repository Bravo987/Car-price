# Outlier

```python

from scipy import stats
```

```python
data.corr()
```

```python
#Checking outlier 
sns.boxplot(x='price',data=data)
```

```python
sns.boxplot(x='horsepower',data=data)
```

```python
sns.boxplot(x='carwidth',data=data)
```

```python
# Removing outlier from horse power
IQR = stats.iqr(data.horsepower,interpolation ='midpoint')
```

```python
print(IQR)
```

```python
# Min and Max limit calculations

#first quartile (Q1)
Q1 = data.horsepower.quantile(0.25)

Q3 = data.horsepower.quantile(0.75)

min_limit= Q1 -1.5*IQR
max_limit= Q3 +1.5*IQR
```

```python
max_limit
```

```python
min_limit
```

```python
data.loc[data['horsepower']>max_limit]
```

# Feature Selection

```python
x=data[['carlength','carwidth','curbweight','enginesize','stroke','compressionratio','horsepower']]
y=data.price
```

```python
x
```

```python
y
```

```python
## Step 2 Checking the correlation
sns.heatmap(data.drop('price',axis=1).corr(),annot=True)
```

# Model Creation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size =0.2)
```

```python
# creating model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_predict = lr.predict(x_test)
```

```python
y_test# actual values
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r2score=r2_score(y_test,y_predict)
```
